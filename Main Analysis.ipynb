# Libraries

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import geopandas as gpd
```

# Questions

```python
# 1. What is the distribution of Airbnb prices in Bangkok?
# 2. How do average prices vary by neighborhood group?
# 3. What is the relationship between room type and price?
# 4. How do the number of reviews and price correlate?
# 5. What are the monthly trends in reviews?
```

# Distribution of Airbnb Prices in Bangkok

```python
# Load dataset

df_listings = pd.read_csv('listings.csv')

# Plot distribution of prices

plt.figure(figsize=(10, 6))
sns.histplot(df_listings['price'], bins=50, kde=True)
plt.title('Distribution of Airbnb Prices in Bangkok')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```

# Avg Prices by Neigbourhood

```python
# Plot average price by neighborhood group
plt.figure(figsize=(12, 6))
sns.barplot(x='neighbourhood_cleansed', y='price', data=df_listings, estimator=np.mean)
plt.title('Average Price by Neighborhood Group in Bangkok')
plt.xlabel('Neighborhood Group')
plt.ylabel('Average Price')
plt.show()
```

# Relationship between Room type and price

```python
# Plot price distribution by room type

# Load dataset
df_listings = pd.read_csv('listings.csv')

# Clean the 'price' column
# Remove non-numeric characters and convert to numeric
df_listings['price'] = df_listings['price'].replace('[\$,]', '', regex=True).astype(float)

# Ensure room_type is treated as a category
df_listings['room_type'] = df_listings['room_type'].astype('category')

# Plot price distribution by room type
plt.figure(figsize=(10, 6))
sns.boxplot(x='room_type', y='price', data=df_listings)
plt.title('Price Distribution by Room Type in Bangkok')
plt.xlabel('Room Type')
plt.ylabel('Price')
plt.show()
```

# Correlation between number of reviews and price

```python
# Scatter plot of number of reviews vs. price
plt.figure(figsize=(10, 6))
sns.scatterplot(x='number_of_reviews', y='price', data=df_listings)
plt.title('Price vs. Number of Reviews in Bangkok')
plt.xlabel('Number of Reviews')
plt.ylabel('Price')
plt.show()
```

# Monthly Trends in Reviews

```python
# Load reviews dataset
df_reviews = pd.read_csv('reviews.csv')

# Convert date to datetime and extract month
df_reviews['date'] = pd.to_datetime(df_reviews['date'])
df_reviews['month'] = df_reviews['date'].dt.to_period('M')

# Group by month and count reviews
reviews_per_month = df_reviews.groupby('month').size()

# Plot monthly trends in reviews
plt.figure(figsize=(12, 6))
reviews_per_month.plot()
plt.title('Monthly Trends in Reviews in Bangkok')
plt.xlabel('Month')
plt.ylabel('Number of Reviews')
plt.show()
```
